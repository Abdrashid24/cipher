```python
import random
```

```python
def unique(it):
    s = set()
    for x in it:
        if x not in s:
            s.add(x)
            yield x

plaintext_xor_e = input("\nНапишите сообщение: ")
alphabet="abcdefghijklmnopqrstuvwxyz"
cipherAscii_e=""
cipherBin_e=""
quanAscii=""

for i in range(0, len(plaintext_xor_e)):
    xor = ord(plaintext_xor_e[i])
    cipherAscii_e = cipherAscii_e + str(xor)
    quanAscii+=str(len(str(xor)))

print("\nПеревод в код ASCII из введенного сообщения:", cipherAscii_e)
print("\nКоличество цифр в букве: ",quanAscii)

cipherBin_e=bin(int(cipherAscii_e))[2:]
print("\nПеревод из полученного ответа в двоичный код: ",cipherBin_e)

key_xor_e=""
for i in range(0,len(cipherBin_e)):
    x=random.randint(0, 1)
    key_xor_e+=str(x)
print("\nКлюч-1: ",key_xor_e)

binary_two_e = ''
index_c = 0
for i in str(cipherBin_e):
    if(i==key_xor_e[index_c]):
        binary_two_e+='0'
    else:
        binary_two_e+='1'
    index_c+=1

binary_ten_e=int(binary_two_e,2)

index_a=0
result_xor=""
binary_text=str(binary_ten_e)
for i in str(binary_text):
    binary_int=int(binary_text[index_a])
    result_xor+=alphabet[binary_int]
    index_a+=1

print("\nИз шифра в двоичный код:", binary_two_e)
print("\nИз двоичного в десятичный:", binary_ten_e)
print("\nИз десятичный в буквы:", result_xor)

plaintext=result_xor
key_m_e = input("\nНапишите Ключ-2: ")
len_p=len(plaintext)
sumt=key_m_e+alphabet
nodup=''.join(unique(sumt))
print("Алфавит         :",alphabet)
print("Алфавит с ключом:",nodup)
cipher=""
num_p=0
index_p=0
for i in range(0,99999):
    if(plaintext[num_p]==alphabet[index_p]):
        cipher+=nodup[index_p]
        num_p+=1
        index_p=-1
        if(num_p==len_p):
            break
    index_p+=1
print("\nГотовый результат: ",cipher)
```
